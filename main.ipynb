```python
import requests
import dns.resolver

# Step 1 - Temp domains load karo
urls = [
    "https://raw.githubusercontent.com/disposable-email-domains/disposable-email-domains/master/disposable_email_blocklist.conf",
    "https://raw.githubusercontent.com/martenson/disposable-email-domains/master/disposable.txt"
]

temp_domains = set()
for url in urls:
    response = requests.get(url)
    temp_domains.update(response.text.splitlines())

extra_domains = ["tempmail.com", "temp-mail.org", "throwmail.com"]
temp_domains.update(extra_domains)

# Step 2 - Email check function
def check_email(email):
    print(f"\nChecking: {email}")
    
    if "@" not in email:
        print("❌ Invalid email - @ nahi hai!")
        return
    
    parts = email.split("@")
    username = parts[0]
    domain = parts[1]
    
    # Temp mail check
    if domain in temp_domains:
        print("❌ Temp mail hai - Block karo!")
        return
    
    # DNS check
    try:
        dns.resolver.resolve(domain, 'MX')
        print("✅ Legit email hai - Allow karo!")
    except:
        print("❌ Fake domain - MX records nahi mile!")

# Test karo
check_email("ajay@google.com")
check_email("ajay@tempmail.com")
check_email("ajay@ajdbasbd.com")
```

```python
import whois
from datetime import datetime

def check_domain_age(domain):
    try:
        w = whois.whois(domain)
        creation_date = w.creation_date
        
        # Kabhi kabhi list aata hai
        if isinstance(creation_date, list):
            creation_date = creation_date[0]
        
        age = datetime.now() - creation_date.replace(tzinfo=None)
        age_days = age.days
        
        print(f"✅ Domain {age_days} din purana hai")
        
        if age_days < 30:
            print("⚠️ Naya domain - Suspicious!")
        else:
            print("✅ Purana domain - Legit lag raha hai!")
            
    except whois.exceptions.WhoisDomainNotFoundError:
        print("❌ Domain exist hi nahi karta!")
    except Exception as e:
        print(f"❌ Kuch aur error aaya: {e}")

check_domain_age("google.com")
check_domain_age("ajdbasbd.com")
```

```python
import requests
import dns.resolver
import whois
from datetime import datetime

# Temp domains load karo
urls = [
    "https://raw.githubusercontent.com/disposable-email-domains/disposable-email-domains/master/disposable_email_blocklist.conf",
    "https://raw.githubusercontent.com/martenson/disposable-email-domains/master/disposable.txt"
]

temp_domains = set()
for url in urls:
    response = requests.get(url)
    temp_domains.update(response.text.splitlines())

extra_domains = ["tempmail.com", "temp-mail.org", "throwmail.com"]
temp_domains.update(extra_domains)

def check_email(email):
    print(f"\nChecking: {email}")
    print("-" * 40)

    # Check 1 - Valid email
    if "@" not in email:
        print("❌ Invalid email - @ nahi hai!")
        return

    parts = email.split("@")
    username = parts[0]
    domain = parts[1]

    # Check 2 - Temp mail
    if domain in temp_domains:
        print("❌ Temp mail hai - Block karo!")
        return

    # Check 3 - DNS/MX check
    try:
        dns.resolver.resolve(domain, 'MX')
        print("✅ MX records valid hain")
    except:
        print("❌ MX records nahi mile - Fake domain!")
        return

    # Check 4 - WHOIS age check
    try:
        w = whois.whois(domain)
        creation_date = w.creation_date
        if isinstance(creation_date, list):
            creation_date = creation_date[0]
        age_days = (datetime.now() - creation_date.replace(tzinfo=None)).days
        print(f"✅ Domain {age_days} din purana hai")
        if age_days < 30:
            print("⚠️ Naya domain - Suspicious!")
        else:
            print("✅ Domain age legit hai")
    except whois.exceptions.WhoisDomainNotFoundError:
        print("❌ Domain exist hi nahi karta!")
        return
    except Exception as e:
        print(f"⚠️ WHOIS check nahi ho paya: {e}")

    print("✅ Email legit hai - Allow karo!")

# Test karo
check_email("ajay@google.com")
check_email("ajay@tempmail.com")
check_email("ajay@ajdbasbd.com")
```

```python
def get_email_score(email):
    score = 0
    print(f"\nChecking: {email}")
    print("-" * 40)

    # Check 1 - Valid email
    if "@" not in email:
        print("❌ Invalid email - @ nahi hai!")
        return

    parts = email.split("@")
    domain = parts[1]

    # Check 2 - Temp mail (+30)
    if domain not in temp_domains:
        score += 30
        print(f"✅ Temp mail nahi hai   +30 | Score: {score}")
    else:
        print(f"❌ Temp mail hai         +0  | Score: {score}")
        print(f"🔴 Final Score: {score}/100 - Block karo!")
        return

    # Check 3 - MX records (+30)
    try:
        dns.resolver.resolve(domain, 'MX')
        score += 30
        print(f"✅ MX records valid     +30 | Score: {score}")
    except:
        print(f"❌ MX records nahi      +0  | Score: {score}")
        print(f"🔴 Final Score: {score}/100 - Block karo!")
        return

    # Check 4 - Domain age (+40)
    try:
        w = whois.whois(domain)
        creation_date = w.creation_date
        if isinstance(creation_date, list):
            creation_date = creation_date[0]
        age_days = (datetime.now() - creation_date.replace(tzinfo=None)).days
        if age_days > 365:
            score += 40
            print(f"✅ Domain purana hai    +40 | Score: {score}")
        else:
            print(f"⚠️ Naya domain          +0  | Score: {score}")
    except:
        print(f"⚠️ WHOIS check failed    +0  | Score: {score}")

    # Final result
    print("-" * 40)
    if score >= 70:
        print(f"🟢 Final Score: {score}/100 - Allow karo!")
    elif score >= 40:
        print(f"🟡 Final Score: {score}/100 - Suspicious!")
    else:
        print(f"🔴 Final Score: {score}/100 - Block karo!")

# Test karo
get_email_score("ajay@google.com")
get_email_score("ajay@tempmail.com")
get_email_score("ajay@ajdbasbd.com")
```

```python
import httpx

def check_website(domain):
    try:
        response = httpx.get(f"http://{domain}", timeout=5, follow_redirects=True)
        content = response.text.lower()
        
        # Parked domain keywords
        parked_keywords = [
            "parked domain",
            "buy this domain",
            "domain for sale",
            "this domain is for sale",
            "search ads"
        ]
        
        for keyword in parked_keywords:
            if keyword in content:
                print(f"⚠️ Parked domain - '{keyword}' mila content mein!")
                return False
        
        print(f"✅ Real website hai!")
        return True
    except:
        print("❌ Website exist nahi karti!")
        return False

check_website("google.com")
check_website("asdad.com")
```

```python
import httpx

def check_website(domain):
    try:
        response = httpx.get(f"http://{domain}", timeout=5, follow_redirects=True)
        content = response.text.lower()

        # 403 = parked/empty domain
        if response.status_code == 403:
            print(f"⚠️ 403 Forbidden - Parked domain!")
            return False

        # Parked domain keywords
        parked_keywords = [
            "parked domain",
            "buy this domain",
            "domain for sale",
            "this domain is for sale",
            "search ads"
        ]

        for keyword in parked_keywords:
            if keyword in content:
                print(f"⚠️ Parked domain - '{keyword}' mila!")
                return False

        print(f"✅ Real website hai!")
        return True
    except:
        print("❌ Website exist nahi karti!")
        return False

check_website("google.com")
check_website("asdad.com")
```
